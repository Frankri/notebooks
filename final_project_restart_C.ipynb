## Simple Linear Regression
As seen in the previous ipython notebooks, there is a correlation between (mean) windspeed and temperature. I will start at this point.

Bxy - regression coefficient,
Rxy - correlation coefficient,
Sx - standard deviation of x,
Sy - standard deviation of y

```python
data[['meanwindspdi', 'meantempi']].corr()
```

```python
data[['meanwindspdi', 'meantempi']].describe()
```

```python
#Values for computation simple linear regression for meantemp

Rxy = -0.3643 # linear regression coefficient 
Sx = 1.982
Sy = 6.568
My = data['meanwindspdi'].mean()
Mx = data['meantempi'].mean()
```

```python
B1 = Rxy * Sx/ Sy # sometimes B1 is called Bxy, b1 
```

```python
B0 = Mx - B1 * My
```

```python
print 'inter:',B0 # Abschnitt, auch alpha 
print 'slope:',B1 # Steigung, auch beta von i 
```

```python
def lin_reg(x):
    y = B0 + B1 * x
    return y
```

```python
def error(y):
    er = (B0 + B1) - y
    return er
```

```python
ctr = 0
arr = []
for e in data.meanwindspdi:
    pt = lin_reg(e)
    arr.append(pt)
    e = er
    #data['predicted_by_windspeed'][ctr] = pt
    ctr += 1
    
print len(arr)
data['t_predicted_by_windspeed'] = arr
```
