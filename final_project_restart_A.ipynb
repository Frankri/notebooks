A closer look in the rainy days:

Let say we like to compare the entries on rainy days with the entries on non rainy days. Can we compare the two data sets?

I will check the entries, because I assume the rain pushes the the people in the subway. In stats translated: H0: The is a difference between enries during rain fall and without rain fall. The H1: No difference between those two weather conditons.

```python
data[['ENTRIESn_hourly', 'EXITSn_hourly', 'rain']].groupby('rain').describe()
```

```python
data[['ENTRIESn_hourly', 'rain']].groupby('rain').hist(bins = 50)
```

```python
rainy_days = data[data.rain == 1]
```

```python
non_rainy_days = data[data.rain == 0]
```

At a first glance: That's not the beloved _Bell Curve_. If we like to get a p-Value we have to perform the Mann-Whitney-U-test for non regular distributions. And at the end we have to multiply with 2 because we did a one sided test.

```python
(u, pvalue) = sp.stats.mannwhitneyu(non_rainy_days["ENTRIESn_hourly"],
                                       rainy_days["ENTRIESn_hourly"])
print "entries on non rainy_days: %s" % non_rainy_days["ENTRIESn_hourly"].median()
print "entries on rainy_days: %s" % rainy_days["ENTRIESn_hourly"].median()
print "p-value of test statistic: %.4f" % (pvalue * 2, )
```

What do we know now? The H1 ist is rejected, what means H0 is correct - but in average 4 people per hour?

__Parenthesis__: Rain or not rain is not the right quality. If we have drizzel nobody cares. In realty that's no rain. Real rain is precipitation more than 1.0 inches per square.

# 5 Linear Regression

The steps below follow [Allen Downey's 'Regression with Python, pandas and StatsModels'](http://allendowney.blogspot.de/2014/09/regression-with-python-pandas-and.html) with __my__ values.

The influence of windspeed on temperature is obvius. I will start with _windspeed_ and _temperature_ for tryout.

```python
windy = data.dropna(subset=['meanwindspdi', 'meantempi'])
```

```python
plt.scatter(data.meanwindspdi, data.meantempi)
```

```python
windy['meanwindspdi'].mean()
```

```python
windy['meantempi'].mean()
```

```python
windy['meanwindspdi'].corr(windy['meantempi'])
```

```python
import statsmodels.formula.api as smf
```

```python
formula = 'meantempi ~ meanwindspdi'
```

```python
results = smf.ols(formula, data = windy).fit()
```

```python
results.summary()
```

R-squared of 0.133 - not so strong (we like to have 0.4 or higher) but ok.
***
That was a test dependency, now I will create the subset for different influences on ridership, that means my dependend variable y are the entries in the subway. My independend features are bundled in x.

I start with __one__ x that's the _temperature_.

```python
entries = data.dropna(subset=['meantempi', 'ENTRIESn_hourly'])
```

```python
entries.head(5)
```

```python
plt.scatter(entries.meantempi, entries.ENTRIESn_hourly)
```

```python
entries['meantempi'].mean()
```

```python
entries['ENTRIESn_hourly'].mean()
```

```python
entries['ENTRIESn_hourly'].corr(entries['meantempi'])
```

```python
formula = 'ENTRIESn_hourly ~ meantempi'
```

```python
results = smf.ols(formula, data = entries).fit()
```

```python
results.summary()
```

Far as I see the mean temperature has not a big influence on the ridership. I will add some other features.
***
Now I will put in the whole bunch of features.

```python
formula = ('ENTRIESn_hourly ~ meantempi + '
            'rain == 1 +'
            'fog == 1 +'
            'precipi +'
            'weekday +'
            'Hour')
```

```python
results = smf.ols(formula, data = entries).fit()
```

```python
results.summary()
```

R-squared: 0.0037 ways too low!
 
One last thing: the turnstiles or better the location of turnstiles - the UNITs. That's a little tricky, because we have _categorial_ data in conntrast to the other (eg. meantempi, meanwindspdi). To solve this problem we have to create so called _dummy variables_. Each value moves to a column and has only 1 or 0. So we created a hughe table with 464 columns and join this with the other features.

```python
dummy_units = pd.get_dummies(data['UNIT'], prefix='unit')
```

```python
dummy_units.head(10)
```

```python
formula = ('ENTRIESn_hourly ~ meantempi + '
            'rain == 1 +'
            'fog == 1 +'
            'precipi +'
            'weekday +'
            'Hour +'
            'dummy_units')
```

```python
results = smf.ols(formula, data = entries).fit()
```

```python
results.summary()
```

Summary: R-squared _0.463_ is good for a prediction. At this point I would say: We solved the __research question__. The features are our influencers and if have we this data we can predict the entries in subway station.

# 7 Open Points - Overfitting, Non-Linearity, Less Data and the Colorblinds

During the work with the data I noted some possible problems. 

### Overfitting
[Overfitting](http://en.wikipedia.org/wiki/Overfitting) means you added to many features in your linear regression model. At the end a weakness will be hide by meaningless features. For the weather data I descided to pull the _rain_ and _fog_ feature out of the set.

### Non-linear Relationships

I found a little trick to boost the data and check if maybe the have a non-linear relationship. It's very easy: square the suspicious data. 

Let's make a try and say the hour isn't linear.

```python
entries['f_non_lin'] = entries['meantempi'] ** 2
```

```python
entries.f_non_lin
```

```python
formula = ('ENTRIESn_hourly ~ meantempi') # hour as single feature
```

```python
results = smf.ols(formula, data = entries).fit()
```

```python
results.summary()
```

```python
formula = ('ENTRIESn_hourly ~ meantempi +'
           'f_non_lin') 
```

```python
results = smf.ols(formula, data = entries).fit()
```

```python
results.summary()
```

No change! Not really helpfull in our case, but maybe good to know.
