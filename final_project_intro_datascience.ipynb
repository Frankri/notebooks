````python
non_rainy_days = data[data.rain == 0]
````

At a first glance: That's not the beloved _Bell Curve_. I would like to get a p-Value and have to perform the Mann-Whitney-U-test for non regular distributions. At the end we will need to _double_ the result because it is a one sided test.

````python
(u, pvalue) = sp.stats.mannwhitneyu(non_rainy_days["ENTRIESn_hourly"],
                                       rainy_days["ENTRIESn_hourly"])
print "entries on non rainy_days: %s" % non_rainy_days["ENTRIESn_hourly"].median()
print "entries on rainy_days: %s" % rainy_days["ENTRIESn_hourly"].median()
print "p-value of test statistic: %.4f" % (pvalue * 2, )
````

What do we know now? The H1 is rejected which means H0 is correct - but in average 4 people per hour?

# 5 Linear Regression

The influence of windspeed on temperature is obvious. I will start with a short example of _windspeed_ and _temperature_.

````python
windy = data.dropna(subset=['meanwindspdi', 'meantempi'])
````

````python
plt.scatter(data.meanwindspdi, data.meantempi)
````

````python
windy['meanwindspdi'].mean()
````

````python
windy['meantempi'].mean()
````

````python
windy['meanwindspdi'].corr(windy['meantempi'])
````

````python
import statsmodels.formula.api as smf
````

````python
formula = 'meantempi ~ meanwindspdi'
````

````python
startTime = time.time()

results = smf.ols(formula, data = windy).fit()

elapsedTime = time.time() - startTime
print "Required time: %.2f seconds" % elapsedTime
````

````python
results.summary()
````

R-squared of 0.133 - not so strong (we like to have 0.4 or higher) but ok for a single feature.
***

I start with __one__ x that's the _temperature_.

Now I will put in a bunch of features.

````python
formula = ('ENTRIESn_hourly ~ meantempi + '
            'rain == 1 +'
            'precipi +'
            'weekday +'
            'Hour')
````

````python
results.summary()
````

R-squared: 0.037 way too low!
 
One last thing: the turnstiles or rather the location of turnstiles - the UNITs. That's a little tricky, because we have _categorial_ data in contrast to the other (e.g. meantempi, meanwindspdi). To solve this problem we have to create so called _dummy variables_. Each value moves to a column and has only 1 or 0. So we created a huge table with 464 columns and join this with the other features.

````python
dummy_units = pd.get_dummies(data['UNIT'], prefix='unit')
````

````python
dummy_units.head(5)
````

````python
entries = data.join(dummy_units)
````

````python
entries.head()
````

````python
formula = ('ENTRIESn_hourly ~ meantempi + '
            'precipi +'
            'weekday +'
            'Hour +'
            'dummy_units')
````

````python
startTime = time.time()

results = smf.ols(formula, data = entries).fit()

elapsedTime = time.time() - startTime
print "Required time: %.2f seconds" % elapsedTime
````

````python
results.summary()
````

Summary: R-squared _0.463_ is good for a prediction. At this point I would say: We solved the __research question__. The features are our influencers and if we have this data we can predict ridership at the subway.

# 6 Map Reduce - a Look on Larger Data Sets
The most power consuming part is computing the _linear squares_ formula (OLS). If we would like to analyze the whole year we need a map reduce algorithm.

In general - What should the map reducer do for us? The _map_ part should split our data in small portions and send they to the reducer. The _reducer_ part should process the pieces. 
I found a library called [Mr. Job (mrjob)](http://mrjob.readthedocs.org/en/latest/guides/why-mrjob.html#overview) to do this on a easy way.


```
from mrjob.job import MRJob

class MRWordFrequencyCount(MRJob):

    def mapper(self, _, line):
        yield "chars", len(line)
        yield "words", len(line.split())
        yield "lines", 1

    def reducer(self, key, values):
        yield key, sum(values)

if __name__ == '__main__':
    MRWordFrequencyCount.run()
```

I started this example with _alice in wonderland.txt_ and got: 
"chars"	144945
"lines"	3598
"words"	26467.

It works!

I have to find a right _mapper_ and _reducer_ parts for my data set.

# 7 Open Points - Overfitting, Non-Linearity, Less Data and the Colorblinds

During the work with the data I noted some possible problems. 

### Overfitting
[Overfitting](http://en.wikipedia.org/wiki/Overfitting) means you added to many features in your linear regression model. At the end a weakness will be obscured by meaningless features. For the weather data I decided to pull the _rain_ and _fog_ features out of the set.

### Non-linear Relationships

I found a little trick to boost the data and check if maybe they have a non-linear relationship. It's very easy: square the suspicious data. 

Let's try and say we guess the _meantempi_ isn't a linear feature.

````python
entries['f_non_lin'] = entries['meantempi'] ** 2 # f_non_lin stands for a non linear feature
````

````python
entries.f_non_lin
````

````python
formula = ('ENTRIESn_hourly ~ meantempi') # temeperature as the non linear feature
````

````python
results = smf.ols(formula, data = entries).fit()
````

````python
results.summary()
````

````python
formula2 = ('ENTRIESn_hourly ~ meantempi +'
           'f_non_lin') 
````

````python
results = smf.ols(formula2, data = entries).fit()
````
