The Enties and Exits grouped by weekdays show we have busy days and not so busy weekends. In the next step I will create a mask with for the busiest hours (7 am and 12pm) and days (Monday till Friday, 0 - 5). At the end I will have a new data frame with the most frequented hours and days.

```python
rush_mask = (data['weekday'] < 5) & (data['Hour'] > 7) & (data['Hour'] < 24)
```

```python
data_rush_hour = pd.DataFrame(data[rush_mask])
```

```python
data_rush_hour.head(5)
```

Lets see what influence the mean temperature on the _rush hour people_ has. I set a treshold at the mean temperature. Below we have the cold days, above the warm. 

```python
low_temp_days = data_rush_hour[data_rush_hour.meantempi < 64]
```

```python
high_temp_days = data_rush_hour[data_rush_hour.meantempi > 64]
```

```python
(u, pvalue) = sp.stats.mannwhitneyu(low_temp_days["ENTRIESn_hourly"],
                                       high_temp_days["ENTRIESn_hourly"])
print "entries on cold days: %s" % low_temp_days["ENTRIESn_hourly"].median()
print "entries on warm days: %s" % high_temp_days["ENTRIESn_hourly"].median()
print "p-value of test statistic: %.4f" % (pvalue * 2, )
```
