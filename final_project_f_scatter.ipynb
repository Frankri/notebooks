## Import the Data in Data Frame

```python
data = pd.read_csv('turnstile_data_master_with_weather.csv')
```

```python
data.columns
```

```python
data.describe()
```

```python
data['DATEn'] = pd.to_datetime(data['DATEn']) # Spalte DATEn in 'echtes' Datum umwandeln    
data['date_delta'] = (data['DATEn'] - data['DATEn'].min())  / np.timedelta64(24,'h') # Datum in float umwandeln
```

```python
data['weekday'] = pd.DatetimeIndex(data['DATEn']).weekday
```

## Tryouts with Random Numbers

```python
rnd = pd.DataFrame(random.randint(10, size = (50, 3))) # creates a data frame with 50 random intergers between 0 and 10 all 3 times (3 columns)
```

```python
rnd.head()
```

```python
rnd2 = rnd.rename(columns = {0: 'eins', 1: 'zwei', 2: 'drei'}) # renaming the column head
```

```python
rnd2.head()
```

```python
ggplot(aes(x = 'eins', y = 'zwei', color = 'drei'), data = rnd2) + geom_point() + stat_smooth(color = 'red') +\
    ggtitle("Random Numbers")
```

***
## Graph with ggplot
Now I use the mta data

```python
ggplot(aes(x='date_delta', y='ENTRIESn_hourly'), data=data) + \
    geom_point(color='lightblue') + \
    stat_smooth(span=.15, color='black', se=True) + \
    ggtitle("Subway Ridership per Day") + \
    xlab("Date") + \
    ylab("Entries")
```

```python
ggplot(aes(x='precipi', y='EXITSn_hourly', color = 'weekday'), data=data) + geom_point() +  stat_smooth(color='red')
```
