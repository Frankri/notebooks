## Import the Data in Data Frame

```python
data = pd.read_csv('20140914-new-spam.csv', sep = ';') #separator!
```

```python
data.columns
```

```python
data2 = data.rename(columns = {'First appearance':'First', 'Last appearance': 'Last', 'Example Subject from sample': 'Subject'}) # rename the column heads
```

```python
data2.head(5)
```

```python
data2['First'] = pd.to_datetime(data2['First']) # Spalte DATEn in 'echtes' Datum umwandeln    
data2['date_delta'] = (data2['First'] - data2['First'].min())  / np.timedelta64(24,'h') # Datum in float umwandeln
```

```python
ggplot(aes(x = 'date_delta', y = 'Count', color = 'Samples'), data = data2) + geom_point() + stat_smooth(color = 'red') +\
    ggtitle("New Spam Samples 9/14/2014")
```

```python
data2.Count.boxplot()
show()
```

```python
data2[['First', 'Last', 'Count', 'Samples']].plot(title ='Entries, Exits and Pressure', x = 'First', subplots=True, figsize=(10,10))
```
